```python
from PyEMD import EMD
import pylab as plt


from pyhht.visualization import plot_imfs
import numpy as np
import math
#from pyhht.emd import EMD
import matplotlib.pyplot as plt
from tftb.generators import fmconst
from scipy import angle, unwrap
from scipy.signal import hilbert
import pandas as pd
from scipy import signal
```

```python
df=pd.read_excel("C:/Users/Dell/Documents/sem8/Data Set/hacat.xlsx")
a1=df["t"]
list(a1)
import numpy as np
a1ar=np.asarray(a1)
type(a1ar)
#print(a1ar)
a2=df["modes"]
list(a2)
import numpy as np
a2ar=np.asarray(a2)
type(a2ar)
#print(a2ar)
x=a1ar+a2ar
plt.plot(a1ar,a2ar);



# Define signal

# Execute EMD on signal
IMF = EMD().emd(a2ar,a1ar)
N = IMF.shape[0]+1
# Plot results
plt.figure(figsize=(18,12))
plt.subplot(N,1,1)
plt.plot(a1ar, a2ar, 'r')
plt.title("Input signal:")
plt.xlabel("Time [s]")
for n, imf in enumerate(IMF):
    plt.figure(figsize=(18,12))
    plt.subplot(N,1,n+2)
    plt.plot(a1ar, imf, 'g')
    plt.title("IMF "+str(n+1))
    plt.xlabel("Time [s]")
plt.tight_layout()
#plt.savefig('simple_example')
plt.show()




hs1 = hilbert(IMF[0])
omega_imf_one = unwrap(angle(hs1))/(2*3.14)
f_inst_imf_one = np.diff(omega_imf_one) 
plt.plot(a1ar[1:658],abs(f_inst_imf_one[1:658])/0.001,'r')

hs2 = hilbert(IMF[1])
omega_imf_2 = unwrap(angle(hs2))/(2*3.14)
f_inst_imf_2 = np.diff(omega_imf_2) 
plt.plot(a1ar[1:658],abs(f_inst_imf_2[1:658])/0.001,'b')

hs3 = hilbert(IMF[2])
omega_imf_3 = unwrap(angle(hs3))/(2*3.14)
f_inst_imf_3 = np.diff(omega_imf_3) 
plt.plot(a1ar[1:658],abs(f_inst_imf_3[1:658])/0.001,'g')

hs4 = hilbert(IMF[3])
omega_imf_4 = unwrap(angle(hs4))/(2*3.14)
f_inst_imf_4 = np.diff(omega_imf_4) 
plt.plot(a1ar[1:658],abs(f_inst_imf_4[1:658])/0.001,'y')
#plt.show()

hs5 = hilbert(IMF[4])
omega_imf_5 = unwrap(angle(hs5))/(2*3.14)
f_inst_imf_5 = np.diff(omega_imf_one) 
plt.plot(a1ar[1:658],abs(f_inst_imf_5[1:658])/0.001,'r')

hs6 = hilbert(IMF[5])
omega_imf_6 = unwrap(angle(hs6))/(2*3.14)
f_inst_imf_6 = np.diff(omega_imf_6) 
plt.plot(a1ar[1:658],abs(f_inst_imf_6[1:658])/0.001,'g')

hs7 = hilbert(IMF[6])
omega_imf_7 = unwrap(angle(hs7))/(2*3.14)
f_inst_imf_7 = np.diff(omega_imf_7) 
plt.plot(a1ar[1:658],abs(f_inst_imf_7[1:658])/0.001,'b')

hs8 = hilbert(IMF[7])
omega_imf_8 = unwrap(angle(hs8))/(2*3.14)
f_inst_imf_8 = np.diff(omega_imf_8) 
plt.plot(a1ar[1:658],abs(f_inst_imf_8[1:658])/0.001,'y')
plt.show()
```
